## Terpedia natural-language API

The client requests safe JSON-serialized Python data. In Colab, store either a personal JWT as `TERPEDIA_TOKEN` or an approved service key as `TERPEDIA_API_KEY` in Secrets. Credentials are read at runtime and are never written into the notebook.


```python
import os, requests
from dataclasses import dataclass

try:
    from google.colab import userdata as colab_userdata
except ImportError:
    colab_userdata = None

def _secret(name):
    if colab_userdata is not None:
        try:
            return colab_userdata.get(name)
        except Exception:
            pass
    return os.environ.get(name)

@dataclass
class TerpediaResult:
    raw: dict

    def __post_init__(self):
        self.variables = self.raw.get("python_data", {}).get("variables", {})
        for name, value in self.variables.items():
            if name.isidentifier() and not hasattr(self, name):
                setattr(self, name, value)

    def frame(self, name="data"):
        value = self.variables.get(name)
        if value is None:
            return pd.DataFrame()
        if isinstance(value, list):
            return pd.json_normalize(value, sep=".")
        if isinstance(value, dict):
            return pd.json_normalize([value], sep=".")
        return pd.DataFrame({name: [value]})

class TerpediaClient:
    def __init__(self, base_url="https://api.terpedia.com", token=None, api_key=None):
        self.base_url = base_url.rstrip("/")
        self.token = token or _secret("TERPEDIA_TOKEN")
        self.api_key = api_key or _secret("TERPEDIA_API_KEY")

    def query(self, question, **options):
        headers = {"Content-Type": "application/json"}
        if self.token:
            headers["Authorization"] = f"Bearer {self.token}"
        elif self.api_key:
            headers["X-API-Key"] = self.api_key
        payload = {"query": question, "response_format": "python_data", **options}
        response = requests.post(f"{self.base_url}/api/v1/query", headers=headers, json=payload, timeout=180)
        response.raise_for_status()
        return TerpediaResult(response.json())

terpedia = TerpediaClient()
research_question = next((p.get("objective") for p in reversed(STUDY["protocols"]) if p.get("objective")), STUDY["study"].get("hypothesis") or STUDY["study"]["title"] )
print("research_question =", research_question)
print("Ready: terpedia_response = terpedia.query(research_question)")
print("Then use terpedia_response.answer, .data, .citations, .variables, or .frame().")
```

## Provenance and artifact verification


```python
from urllib.request import urlopen

def verify_https_artifact(record):
    uri = record.get("uri", "")
    expected = record.get("sha256")
    if not uri.startswith("https://"):
        return {"uri": uri, "status": "credential_or_protocol_required", "expected_sha256": expected}
    with urlopen(uri) as response:
        actual = hashlib.sha256(response.read()).hexdigest()
    return {"uri": uri, "status": "verified" if actual == expected else "checksum_mismatch", "expected_sha256": expected, "actual_sha256": actual}

artifact_manifest = [{k: item.get(k) for k in ("id", "name", "uri", "sha256", "role")} for item in STUDY["artifacts"]]
display(pd.DataFrame(artifact_manifest))
print("Call verify_https_artifact(record) for public HTTPS artifacts. gs:// artifacts require authenticated Google Cloud access.")
```
