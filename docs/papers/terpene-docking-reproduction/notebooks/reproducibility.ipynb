## Terpedia natural-language API

The client requests safe JSON-serialized Python data. In Colab, store either a personal JWT as `TERPEDIA_TOKEN` or an approved service key as `TERPEDIA_API_KEY` in Secrets. Credentials are read at runtime and are never written into the notebook.


```python
import os, requests
from dataclasses import dataclass

try:
    from google.colab import userdata as colab_userdata
except ImportError:
    colab_userdata = None

def _secret(name):
    if colab_userdata is not None:
        try:
            return colab_userdata.get(name)
        except Exception:
            pass
    return os.environ.get(name)

@dataclass
class TerpediaResult:
    raw: dict

    def __post_init__(self):
        self.variables = self.raw.get("python_data", {}).get("variables", {})
        for name, value in self.variables.items():
            if name.isidentifier() and not hasattr(self, name):
                setattr(self, name, value)

    def frame(self, name="data"):
        value = self.variables.get(name)
        if value is None:
            return pd.DataFrame()
        if isinstance(value, list):
            return pd.json_normalize(value, sep=".")
        if isinstance(value, dict):
            return pd.json_normalize([value], sep=".")
        return pd.DataFrame({name: [value]})

class TerpediaClient:
    def __init__(self, base_url="https://api.terpedia.com", token=None, api_key=None):
        self.base_url = base_url.rstrip("/")
        self.token = token or _secret("TERPEDIA_TOKEN")
        self.api_key = api_key or _secret("TERPEDIA_API_KEY")

    def query(self, question, **options):
        headers = {"Content-Type": "application/json"}
        if self.token:
            headers["Authorization"] = f"Bearer {self.token}"
        elif self.api_key:
            headers["X-API-Key"] = self.api_key
        payload = {"query": question, "response_format": "python_data", **options}
        response = requests.post(f"{self.base_url}/api/v1/query", headers=headers, json=payload, timeout=180)
        response.raise_for_status()
        return TerpediaResult(response.json())

terpedia = TerpediaClient()
research_question = next((p.get("objective") for p in reversed(STUDY["protocols"]) if p.get("objective")), STUDY["study"].get("hypothesis") or STUDY["study"]["title"] )
print("research_question =", research_question)
print("Ready: terpedia_response = terpedia.query(research_question)")
print("Then use terpedia_response.answer, .data, .citations, .variables, or .frame().")
```

## Provenance and artifact verification


```python
from urllib.request import urlopen

def verify_https_artifact(record):
    uri = record.get("uri", "")
    expected = record.get("sha256")
    if not uri.startswith("https://"):
        return {"uri": uri, "status": "credential_or_protocol_required", "expected_sha256": expected}
    with urlopen(uri) as response:
        actual = hashlib.sha256(response.read()).hexdigest()
    return {"uri": uri, "status": "verified" if actual == expected else "checksum_mismatch", "expected_sha256": expected, "actual_sha256": actual}

artifact_manifest = [{k: item.get(k) for k in ("id", "name", "uri", "sha256", "role")} for item in STUDY["artifacts"]]
display(pd.DataFrame(artifact_manifest))
print("Call verify_https_artifact(record) for public HTTPS artifacts. gs:// artifacts require authenticated Google Cloud access.")
```

### Target-fishing consensus and tool dependence


```python
import re, unicodedata

def normalized_name(value):
    value = str(value or "").replace("α", "alpha").replace("β", "beta").replace("γ", "gamma")
    value = unicodedata.normalize("NFKD", value).encode("ascii", "ignore").decode("ascii").casefold()
    return re.sub(r"[^a-z0-9]", "", value)

def stable_target_key(row):
    return row.get("chembl_id") or row.get("uniprot_id") or normalized_name(row.get("target_name"))

for frame in (target_predictions, author_consensus):
    frame["ligand_key"] = frame["terpene"].map(normalized_name)
    frame["target_key"] = frame.apply(stable_target_key, axis=1)

def consensus_pairs(frame, omitted_tool=None, minimum_tools=2):
    current = frame if omitted_tool is None else frame[frame["tool"] != omitted_tool]
    grouped = current.groupby(["ligand_key", "target_key"], as_index=False)["tool"].agg(lambda values: sorted(set(values))).rename(columns={"tool": "tools"})
    grouped["tool_count"] = grouped["tools"].map(len)
    return grouped[grouped["tool_count"] >= minimum_tools].reset_index(drop=True)

recomputed_consensus = consensus_pairs(target_predictions)
author_pairs = set(map(tuple, author_consensus[["ligand_key", "target_key"]].drop_duplicates().to_numpy()))
recomputed_pairs = set(map(tuple, recomputed_consensus[["ligand_key", "target_key"]].to_numpy()))
assert len(author_pairs) == 257
consensus_comparison = pd.Series({
    "recomputed_pairs": len(recomputed_pairs), "author_pairs": len(author_pairs),
    "overlap": len(recomputed_pairs & author_pairs),
    "precision_vs_author": len(recomputed_pairs & author_pairs) / len(recomputed_pairs),
    "recall_vs_author": len(recomputed_pairs & author_pairs) / len(author_pairs),
}, name="value").to_frame()
display(consensus_comparison)

tools = ["SwissTargetPrediction", "SEA", "STarFish", "PPB2"]
ablation = []
for omitted in tools:
    reduced = consensus_pairs(target_predictions, omitted_tool=omitted)
    pairs = set(map(tuple, reduced[["ligand_key", "target_key"]].to_numpy()))
    ablation.append({"omitted_tool": omitted, "consensus_pairs": len(pairs), "pairs_lost": len(recomputed_pairs - pairs), "author_recall": len(pairs & author_pairs) / len(author_pairs)})
leave_one_tool_out = pd.DataFrame(ablation).sort_values("consensus_pairs", ascending=False)
display(leave_one_tool_out)
display(target_frequencies[target_frequencies["selected_top_five"]][["rank", "target_name", "terpenes_predicting"]])
```

### TF labels, ligand efficiency, and molecular-size effects


```python
TARGET_ALIASES = {
    "cannabinoidreceptor2": "CB2", "cannabinoidcb2receptor": "CB2", "cb2": "CB2",
    "androgenreceptor": "AR", "ar": "AR", "cytochromep45019a1": "CYP19A1", "aromatase": "CYP19A1", "cyp19a1": "CYP19A1",
    "estrogenreceptorbeta": "ERbeta", "erbeta": "ERbeta", "11betahydroxysteroiddehydrogenase1": "11beta-HSD1", "11betahsd1": "11beta-HSD1", "bhd": "11beta-HSD1",
}
def target_key(value):
    key = normalized_name(value)
    return TARGET_ALIASES.get(key, key)

author_tf_pairs = {(normalized_name(row.terpene), target_key(row.target_name)) for row in author_consensus.itertuples()}
def decorate_docking(frame, method):
    result = frame.copy()
    result["method"] = method
    result["tf_positive"] = [
        (normalized_name(row.ligand), target_key(row.target)) in author_tf_pairs for row in result.itertuples()
    ]
    result["ligand_efficiency"] = result["score_kcal_mol"].abs() / result["heavy_atoms"]
    result["score_magnitude"] = result["score_kcal_mol"].abs()
    return result

docking = pd.concat([decorate_docking(moe_docking, "MOE GBVI/WSA dG"), decorate_docking(vina_docking, "AutoDock Vina")], ignore_index=True)

correlation_rows = []
for (method, target), group in docking.groupby(["method", "target"]):
    known = group[group["ligand_class"] == "known_ligand"]
    for label, ligand_class in (("monoterpenes", "monoterpene"), ("sesquiterpenes", "sesquiterpene")):
        terpene_class = group[group["ligand_class"] == ligand_class]
        correlation_rows.append({
            "method": method, "target": target, "series": label,
            "terpenes_plus_known": pd.concat([terpene_class, known])[["heavy_atoms", "score_magnitude"]].corr().iloc[0, 1],
            "terpenes_only": terpene_class[["heavy_atoms", "score_magnitude"]].corr().iloc[0, 1],
        })
size_correlations = pd.DataFrame(correlation_rows)
display(size_correlations)

ligand_efficiency_summary = docking.groupby(["method", "target", "ligand_class"])["ligand_efficiency"].agg(["count", "mean", "median", "min", "max"]).reset_index()
display(ligand_efficiency_summary)
```

### Target-fishing enrichment by docking rank

As in the reference paper, known ligands and TF-positive terpenes are counted as positives. Undefined one-class cases remain explicitly undefined. Values are descriptive because each target contains only 23 terpenes.


```python
import math
def enrichment_curve(population):
    population = population.sort_values("score_kcal_mol").reset_index(drop=True)
    positive = (population["ligand_class"] == "known_ligand") | population["tf_positive"]
    identifiable = 0 < int(positive.sum()) < len(population)
    rows = []
    for percent in range(10, 101, 10):
        top_n = max(1, math.ceil(len(population) * percent / 100))
        factor = (positive.iloc[:top_n].mean() / positive.mean()) if identifiable else None
        rows.append({"percent": percent, "top_n": top_n, "enrichment_factor": factor, "identifiable": identifiable})
    return rows

enrichment_rows = []
for (method, target), group in docking.groupby(["method", "target"]):
    known = group[group["ligand_class"] == "known_ligand"]
    for label, ligand_class in (("monoterpenes", "monoterpene"), ("sesquiterpenes", "sesquiterpene")):
        population = pd.concat([known, group[group["ligand_class"] == ligand_class]], ignore_index=True)
        for row in enrichment_curve(population):
            enrichment_rows.append({"method": method, "target": target, "series": label, **row})
enrichment = pd.DataFrame(enrichment_rows)
display(enrichment)

fig, axes = plt.subplots(2, 2, figsize=(13, 9), sharex=True)
for row_index, method in enumerate(["MOE GBVI/WSA dG", "AutoDock Vina"]):
    for column_index, series in enumerate(["monoterpenes", "sesquiterpenes"]):
        ax = axes[row_index, column_index]
        subset = enrichment[(enrichment["method"] == method) & (enrichment["series"] == series)]
        for target, points in subset.groupby("target"):
            ax.plot(points["percent"], points["enrichment_factor"], marker="o", label=target)
        ax.axhline(1, color="#666", linestyle="--", linewidth=1)
        ax.set_title(f"{method} | {series}")
        ax.set_ylabel("Enrichment factor")
        ax.set_xlabel("Top-ranked fraction (%)")
axes[0, 0].legend(fontsize=8)
fig.tight_layout()
plt.show()
```

### MOE-Vina rank concordance


```python
concordance_rows = []
for target in sorted(set(moe_docking["target"]) & set(vina_docking["target"])):
    shared = moe_docking[moe_docking["target"] == target][["ligand", "score_kcal_mol"]].merge(
        vina_docking[vina_docking["target"] == target][["ligand", "score_kcal_mol"]], on="ligand", suffixes=("_moe", "_vina")
    )
    moe_top = set(shared.nsmallest(5, "score_kcal_mol_moe")["ligand"])
    vina_top = set(shared.nsmallest(5, "score_kcal_mol_vina")["ligand"])
    concordance_rows.append({
        "target": target, "shared_ligands": len(shared),
        "spearman_rho": shared[["score_kcal_mol_moe", "score_kcal_mol_vina"]].corr(method="spearman").iloc[0, 1],
        "top_5_overlap": len(moe_top & vina_top), "top_5_jaccard": len(moe_top & vina_top) / len(moe_top | vina_top),
    })
method_concordance = pd.DataFrame(concordance_rows)
display(method_concordance)
```
